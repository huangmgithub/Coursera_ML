```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.optimize as opt
from sklearn.metrics import classification_report
import scipy.io as sio  # 加载matlab格式的数据集
```

```python
def load_data(path):
    data = sio.loadmat(path)
    X = data['X']
    y = data['y'].flatten()
    return X,y
```

```python
path = './ex3data1.mat'
X, y = load_data(path)
X.shape, y.shape
```

```python
def plot_100_images(X):
    samples = np.random.choice(np.arange(5000), 100)
    images = X[samples]
    
    fig, ax = plt.subplots(nrows = 10, ncols = 10, sharex = True, sharey = True, figsize=(8,8))
    
    for row in range(10):
        for col in range(10):
            ax[row, col].matshow(images[10*row + col].reshape((20, 20)), cmap='gray_r')
    
    plt.xticks([])
    plt.yticks([])
    plt.show()
```

```python
plot_100_images(X)
```
