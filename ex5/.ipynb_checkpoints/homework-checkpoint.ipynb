# Regularized Linear Regression 正则线性回归

## Visualizing the dataset

我们将从可视化数据集开始，其中包含水位变化的历史记录，x，以及从大坝流出的水量，y。

这个数据集分为了三个部分：
+ training set 训练集：训练模型
+ cross validation set 交叉验证集：选择正则化参数(or 多项式次数)
+ test set 测试集：评估性能，模型训练中不曾用过的样本

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import  scipy.io as sio
```

```python
data = sio.loadmat('./ex5/ex5data1.mat')
# 训练集
X, y = data['X'], data['y']
# 验证集
Xval, yval = data['Xval'], data['yval']
# 测试集
Xtest, ytest = data['Xtest'], data['ytest']
```

```python
# 插入偏置项
X = np.insert(X, 0, 1, axis=1)
Xval = np.insert(Xval, 0, 1, axis=1)
Xtest = np.insert(Xtest, 0, 1, axis=1)
print("X={},y={}".format(X.shape, y.shape))
print("Xval={},yval={}".format(Xval.shape, yval.shape))
print('Xtest={},ytest={}'.format(Xtest.shape, ytest.shape))
```

```python
def plot_data():
    plt.figure(figsize=(8,5))
    plt.scatter(X[:,1:], y, c='r', marker='x')
    plt.xlabel('Change in water level (x)')
    plt.ylabel('Water flowing out of the dam (y)')
    plt.grid(True)
    
plot_data()
```

## Regularized linear regression cost function

$$ J(\theta) =  \frac{1}{2m} (\sum_{i=1}^m(h_\theta(x^{(i)}) - y^{(i)})^2) + \frac{\lambda}{2m}\sum_{j=1}^n \theta_j^2 $$ 

```python
theta = np.ones(X.shape[1])
```

```python
# 正则化线性回归代价函数
def cost_reg(theta, X, y, l):
    '''
    do not regularizethe theta0
    theta is a 1-d array with shape (n+1,)
    X is a matrix with shape (m, n+1)
    y is a matrix with shape (m, 1)
    '''
    cost = ((X @ theta - y.flatten()) ** 2).sum()
    reg = l * (theta[1:] @ theta[1:])
    return (cost + reg) / (2 * len(X))
```

```python
print(cost_reg(theta, X, y, l=1))
```

## Regularized linear regression gradient

$$ \frac{\partial }{\partial \theta_0}J(\theta) = \frac{1}{m} \sum_{i=1}^m (h_\theta(x^{(i)}) - y^{(i)})x_j{(i)} \qquad if \quad j = 0 $$ 
$$ \frac{\partial }{\partial \theta_j}J(\theta) = (\frac{1}{m} \sum_{i=1}^m (h_\theta(x^{(i)}) - y^{(i)})x_j{(i)}) + \frac{\lambda}{m}\theta_j  \qquad if \quad j >= 1 $$  

```python
# 梯度
def gradient_reg(theta, X, y, l):
    """
    theta: 1-d array with shape (2,)
    X: 2-d array with shape (12, 2)
    y: 2-d array with shape (12, 1)
    l: lambda constant
    grad has same shape as theta (2,)
    """
    grad = (X @ theta - y.flatten()) @ X
    reg = l * theta
    theta[0] = 0
    return (grad + reg) / len(X)
print(gradient_reg(theta, X, y, 1))
```

## Fitting linear regression 拟合线性回归

```python
import scipy.optimize as opt

def train_linear_reg(X, y, l=0):
    theta = np.zeros(X.shape[1])
    res = opt.minimize(fun=cost_reg,  # 代价函数
                       x0=theta,  # 初始化参数
                       args=(X, y ,l), 
                       method='TNC', 
                       jac=gradient_reg) # 梯度
    return res.x

final_theta = train_linear_reg(X, y,l=0)
```

```python
def plot_data_2():
    fig = plt.figure(figsize=(8,5))
    ax = plt.axes()
    
    ax.scatter(X[:,1:], y, c='r', marker='x')
    ax.plot(X[:,1], X @ final_theta)
    ax.set_xlabel('Change in water level (x)')
    ax.set_ylabel('Water flowing out of the dam (y)')
    ax.grid(True)
    
    plt.show()
plot_data_2()
```

这里我们把λ = 0，因为我们现在实现的线性回归只有两个参数，这么低的维度，正则化并没有用。

从图中可以看到，拟合最好的这条直线告诉我们这个模型并不适合这个数据。

在下一节中，您将实现一个函数来生成学习曲线，它可以帮助您调试学习算法，即使可视化数据不那么容易。

# Bias-variance

机器学习中一个重要的概念是偏差（bias）和方差（variance）的权衡。高偏差意味着欠拟合，高方差意味着过拟合。

在这部分练习中，您将在学习曲线上绘制训练误差和验证误差，以诊断bias-variance问题。

## Learning curves 学习曲线

$$ J_{train}(\theta) =  \frac{1}{2m} (\sum_{i=1}^m(h_\theta(x^{(i)}) - y^{(i)})^2) $$

训练样本X从1开始逐渐增加，训练出不同的参数向量θ。接着通过交叉验证样本Xval计算验证误差。

+ 使用训练集的子集来训练模型，得到不同的theta。

+ 通过theta计算训练代价和交叉验证代价，切记此时不要使用正则化，将 λ=0 。

+ 计算交叉验证代价时记得整个交叉验证集来计算，无需分为子集。

```python
# 绘制学习曲线
def plot_learning_curve(X, y, Xval, yval, l):
    size = range(1, len(X) + 1)
    train_cost = []
    cv_cost = []
    for i in size:
        fit_theta = train_linear_reg(X[:i], y[:i], l)
        train_cots_i = cost_reg(fit_theta, X[:1], y[:i], l)
        cv_cost_i = cost_reg(fit_theta, Xval, yval, l)
        train_cost.append(train_cots_i)
        cv_cost.append(cv_cost_i)
    
    fig = plt.figure()
    ax = plt.axes()
    ax.plot(size, train_cost, c='r', label='training cost')
    ax.plot(size, cv_cost, c='b', label='cv cost')
    ax.legend()
    ax.set_xlabel('training examples')
    ax.set_ylabel('error')
    ax.set_title('Learning curve for linear regression')
    ax.grid(True)
    
    plt.show()
plot_learning_curve(X, y, Xval, yval, l=0)
```

从图中看出来，随着样本数量的增加，训练误差和交叉验证误差都很高，这属于高偏差，欠拟合????????怎么绘制成这样了。。。

# Polynomial regression 多项式回归

我们的线性模型对于数据来说太简单了，导致了欠拟合(高偏差)。在这一部分的练习中，您将通过添加更多的特性来解决这个问题。

使用多项式回归，假设函数形式如下：

$$ h_\theta(x) = \theta_0 + \theta_1 * (waterLevel) + \theta_2 * (waterLevel)^2 + ... + \theta_p * (waterLevel)^p= \theta_0 + \theta_1x_1 + \theta_2x_2 + ... + \theta_px_p $$

## Learning Polynomial Regression

数据预处理

+ X，Xval，Xtest都需要添加多项式特征，这里我们选择增加到6次方，因为若选8次方无法达到作业pdf上的效果图，因为scipy和octave版本的优化算法不同。

+ 不要忘了标准化

```python
def get_poly_features(X, power):
    """获得多项式特征"""
    Xpoly = X.copy()
    for i in range(2, power + 1):
        Xpoly = np.insert(Xpoly, Xpoly.shape[1], np.power(Xpoly[:,1], i), axis=1)
    return Xpoly

def get_means_stds(X):
    """获得均值和标准差"""
    means = np.mean(X, axis=0)
    stds = np.std(X, axis=0, ddof=1) # ddof=1 means 样本标准差
    return means, stds

def normalize_features(X, means, stds):
    """归一化"""
    Xnorm = X.copy()
    Xnorm[:, 1:] = Xnorm[:, 1:] - means[1:]
    Xnorm[:, 1:] = Xnorm[:, 1:] / stds[1:]
    return Xnorm
    
```

##### 关于归一化，所有数据集应该都用训练集的均值和样本标准差处理。切记。所以要将训练集的均值和样本标准差存储起来，对后面的数据进行处理。

而且注意这里是样本标准差而不是总体标准差，使用np.std()时，将ddof=1则是样本标准差，默认=0是总体标准差。而pandas默认计算样本标准差。

获取添加多项式特征以及 标准化之后的数据

```python
power = 6
train_means, train_stds = get_means_stds(get_poly_features(X, power))
X_norm = normalize_features(get_poly_features(X, power), train_means, train_stds)
Xval_norm = normalize_features(get_poly_features(Xval, power), train_means, train_stds)
Xtest_norm = normalize_features(get_poly_features(Xtest, power), train_means, train_stds)
X_norm.shape
```

```python
def plot_fit(means, stds, l):
    theta = train_linear_reg(X_norm, y, l)
    x = np.linspace(-75,55,50)
    x_mat = x.reshape(-1,1)
    x_mat = np.insert(x_mat, 0, 1, axis=1)
    x_mat = get_poly_features(x_mat, power)
    x_mat_norm = normalize_features(x_mat, means, stds)
    
    plot_data()
    plt.plot(x, x_mat_norm @ theta, 'b--')
```

```python
plot_fit(train_means, train_stds, 0)
```

```python
plot_learning_curve(X_norm, y, Xval_norm, yval, 0)
```

## Adjusting the regularization parameter

上图可以看到 λ  = 0时，训练误差太小了，明显过拟合了。

我们继续调整λ  = 1 时

```python
plot_fit(train_means, train_stds, 1)
plot_learning_curve(X_norm, y, Xval_norm, yval, 1)
```

我们继续调整λ = 100 时，很明显惩罚过多，欠拟合了

```python
plot_fit(train_means, train_stds, 100)
plot_learning_curve(X_norm, y, Xval_norm, yval, 100)
```

## Selecting λ using a cross validation set

```python
lambdas = [0., 0.001, 0.003, 0.01, 0.03, 0.1, 0.3, 1., 3., 10.]
errors_train = []
errors_cv = []
for l in lambdas:
    theta = train_linear_reg(X_norm, y, l)
    errors_train.append(cost_reg(theta, X_norm, y, l))
    errors_cv.append(cost_reg(theta, Xval_norm, yval, l))

plt.figure(figsize=(9,5))
plt.plot(lambdas, errors_train, c='r', label='Train')
plt.plot(lambdas, errors_cv, c='b', label='Cross Validation')
plt.legend()
plt.xlabel('lambda')
plt.ylabel('Error')
plt.grid(True)
```

```python
# 可以看到时交叉验证代价最小的是 lambda = 3
lambdas[np.argmin(errors_cv)]  # 3.0 ???????
```
