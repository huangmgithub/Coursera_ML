```python
def plot_boundary(clf ,X):
    """plot decision boundary"""
    x_min, x_max = X[:,0].min() * 1.2, X[:,0].max() * 1.1  # 为了扩充边界
    y_min, y_max = X[:,1].min() * 1.1, X[:,1].max() * 1.1  # 为了扩充边界
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 500), np.linspace(y_min, y_max, 500))
    Z = clf.predict(np.c_[xx.flatten(), yy.flatten()])
    Z = Z.reshape(xx.shape)
    plt.contour(xx, yy, Z) # 等高线
```

```python
models = [svm.SVC(C, kernel='linear') for C in [1, 100]]
clfs = [model.fit(X, y.flatten()) for model in models]
```

```python
titles = ['SVM Decision Boundary with C = {} (Example Dataset 1'.format(C) for C in [1, 100]]
for clf, title in zip(clfs, titles):
    plt.figure(figsize=(8,5))
    plot_data(X, y)
    plot_boundary(clf,X)
    plt.title(title)
```

可以从上图看到：  
当C比较小时，对应的正则化参数\lambda较大， 对参数的惩罚增大，属于高偏差，比较宽松，允许一定的误分类存在，间隔较大。  
当C比较大时，对应的正则化参数\lambda较小， 对参数的惩罚增小，拟合很好，比较严格，误分类少，但间隔比较狭窄。  

## SVM with Gaussian Kernels

使用SVM做非线性分类。我们将使用高斯核函数。
为了用SVM找出一个非线性的决策边界，我们首先要实现高斯核函数。我可以把高斯核函数想象成一个相似度函数，用来测量一对样本的距离，(x^(i), y^(i) )

$$ K_gaussian(x^{(i)},x^{(j)}) = exp(-\frac{\lvert x^{(i)} - x^{(j)}\rvert^2}{2\sigma^2}) = exp(-\frac{\sum_{k=1}^n(x_k^{(i)} - x_k^{(j)})^2}{2\sigma^2})$$

使用sklearn自带的svm中的核函数即可

### Gaussian Kernel

```python
def gauss_kernel(x1, x2, sigma):
    """高斯核函数"""
    return np.exp(-((x1 - x2) ** 2).sum() / (2 * (sigma ** 2)))
gauss_kernel(np.array([1, 2, 1]),np.array([0, 4, -1]), 2.) 
```

```python
Cvalues = (0.01, 0.03, 0.1, 0.3, 1., 3., 10., 30.)
Sigmavalues = Cvalues
best_C_Sigma = ()
best_score = 0

for C in Cvalues:
    for sigma in Sigmavalues:
        gamma = np.power(sigma, -2) / 2
        model = svm.SVC(C=C, kernel='rbf', gamma=gamma)
        model.fit(X3, y3.flatten())
        current_score = model.score(Xval, yval)
        if current_score > best_score:
            best_score = current_score
            best_C_Sigma = (C, sigma)
print('Best C={}, Best_sigma={}, Best score={}'.format(best_C_Sigma[0], best_C_Sigma[1], best_score))
```

# Spam Classification

这部分用SVM建立一个垃圾邮件分类器。你需要将每个email变成一个n维的特征向量，这个分类器将判断给定一个邮件x是垃圾邮件(y=1)或不是垃圾邮件(y=0)

```python
with open('./ex6/emailSample1.txt') as f:
    email = f.read()
    print(email)
```

可以看到，邮件内容包含 a URL, an email address(at the end), numbers, and dollar amounts. 很多邮件都会包含这些元素，但是每封邮件的具体内容可能会不一样。因此，处理邮件经常采用的方法是标准化这些数据，把所有URL当作一样，所有数字看作一样。

例如，我们用唯一的一个字符串‘httpaddr’来替换所有的URL，来表示邮件包含URL，而不要求具体的URL内容。这通常会提高垃圾邮件分类器的性能，因为垃圾邮件发送者通常会随机化URL，因此在新的垃圾邮件中再次看到任何特定URL的几率非常小。

我们可以做如下处理：

1. Lower-casing: 把整封邮件转化为小写。
2. Stripping HTML: 移除所有HTML标签，只保留内容。
3. Normalizing URLs: 将所有的URL替换为字符串 “httpaddr”.
4. Normalizing Email Addresses: 所有的地址替换为 “emailaddr”
5. Normalizing Dollars: 所有dollar符号($)替换为“dollar”.
6. Normalizing Numbers: 所有数字替换为“number”
7. Word Stemming(词干提取): 将所有单词还原为词源。例如，“discount”, “discounts”, “discounted” and “discounting”都替换为“discount”。
8. Removal of non-words: 移除所有非文字类型，所有的空格(tabs, newlines, spaces)调整为一个空格.

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.io as sio
from sklearn import svm
import re
```

```python
def process_email(email):
    """email preprocessing"""
    email = email.lower() # 转换为小写
    email = re.sub('<[^<>]>', '', email) # 移除所有HTML标签
    email = re.sub('(http|https)://[^\s]*', 'httpaddr', email) # 所有的URL替换为字符串 “httpaddr”.
    email = re.sub('[^\s]+@[^\s]+', 'emailaddr', email) # 所有的地址替换为 “emailaddr”
    email = re.sub('\$','dollar', email)
    email = re.sub('\d+', 'number', email)
    return email
```

```python
# NLP包
import nltk, nltk.stem.porter

def get_token_list(email):
    stemmer = nltk.stem.porter.PorterStemmer()
    
    email = process_email(email)
    
    # 将邮件分割为单个单词，re.split() 可以设置多种分隔符
    tokens = re.split('[ \@\$\/\#\.\-\:\&\*\+\=\[\]\?\!\(\)\{\}\,\'\"\>\_\<\;\%]', email)
    
    tokenlist = []
    for token in tokens:
        # 删除任何非字母数字的字符
        token = re.sub('[^a-zA-Z0-9]','',token)
        # Use the Porter stemmer to 提取词根
        stemmed = stemmer.stem(token)
        if not len(token):continue
        tokenlist.append(stemmed)
    return tokenlist
```

### Vocabulary List(词汇表)

```python
def get_vocab_indices(email, vocab):
    """计算出email中存在多少与vocab一样的单词"""
    tokens = get_token_list(email)
    index = [i for i in range(len(vocab)) if vocab[i] in tokens]
    return index
```

```python
def get_feature_vector(email):
    """
    将email转化为词向量，n是vocab的长度。存在单词的相应位置的值置为1，其余为0
    """
    df = pd.read_table('./ex6/vocab.txt', names=['words'])
    vocab = df.as_matrix().flatten()
    vector = np.zeros(len(vocab))
    index = get_vocab_indices(email, vocab)
    for i in index:
        vector[i] = 1
    return vector
```
